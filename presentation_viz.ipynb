```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from ICA import GeneralizedExponential

```

```python
tf.random.set_seed(42)
```

```python
# Create range of values for a
a = np.linspace(-4, 4, 1000)
R_values = [0.5, 1.0, 2.0, 4.0]
```

```python
m = 1
beta =1
```

```python
a = tf.constant(a[:np.newaxis], dtype=tf.float32)
# define plot
plt.figure(figsize=(12,6))

for R in R_values:
    g = GeneralizedExponential(m)
    # Set R and beta values
    g.R.assign([R])
    g.beta.assign([beta])
    # Compute density
    density = g.pdf(a)
    
    # Plot
    plt.plot(a, density, label=f'R = {R}', linewidth=2)
    
plt.title(f'Generalized Exponential Density (β = {beta})')
plt.xlabel('a')
plt.ylabel('G(a; R, β)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
